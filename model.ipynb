```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder,Normalizer
from sklearn.ensemble import RandomForestRegressor
print(pd.__version__)
print(np.__version__) 
print(sns.__version__)
```

```python
# Load dataset
df = pd.read_csv('Bengaluru_House_Data.csv')
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
#Dropping null values
df=df.drop(['area_type','society','balcony','availability'],axis=1)
```

```python
df.info()
```

```python
#handling null values
df['location'].value_counts()
df['location']=df['location'].fillna('Whitefield') #fill the missing values with most frequent value
df.info()
```

```python
#handlinhg null values in size
df['size'].value_counts()
df['size']=df['size'].fillna('2 BHK') #fill the missing values with most frequent value
df.info()
```

```python
#handling null values in bath
df['bath']=df['bath'].fillna(df['bath'].median()) 
df.info()
```

```python
#cleaning 'size' column
df['size'].unique()
out=[i.split()[0] for i in df['size']]
df['bhk']=out
df['bhk']=df['bhk'].astype(int)
df.info()
```

```python
df=df.drop('size', axis=1)
```

```python
df['total_sqft'].unique()

def convert(a):
    l=a.split('-')
    if len(l)==2:
        return (float(l[0])+float(l[1]))/2
    try:
        return float(a)
    except:
        return None
df['total_sqft']=df['total_sqft'].apply(convert)
df['total_sqft']=df['total_sqft'].fillna(df['total_sqft'].median())

    
```

```python
df['price_per_sqft']=df['price']*100000/df['total_sqft']
```

```python
df
```

```python
#cleaning location
df['location'].value_counts()
```

```python
location_count=df['location'].value_counts()
location_less_than_10=location_count[location_count<=10]
location_less_than_10
```

```python
df['location']=df['location'].apply(lambda x: 'other' if x in location_less_than_10 else x)
df['location'].value_counts()
```

```python
df=df[df['total_sqft']/df['bhk']>=300]
df.describe()
```

```python
df=df[df['bath']<=df['bhk']]
df.info()
df.describe()
```

```python
df=df[df['bhk']<=6] #removing outliers where bhk>6
df.describe()
```

```python
sns.boxplot(y='price_per_sqft', data=df)
plt.show()
```

```python
#handling outliers in price_per_sqft
q1=df['price_per_sqft'].quantile(0.25)
q3=df['price_per_sqft'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-0.5*iqr
upper_bound=q3+0.5*iqr
df=df[(df['price_per_sqft']>=lower_bound) & (df['price_per_sqft']<=upper_bound)]
sns.boxplot(y='price_per_sqft', data=df)
```

```python
df.info()
```

```python
df=df.drop('price_per_sqft', axis=1)
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df=df.drop('index',axis=1)
df
```

```python
copied_dataset=df.copy()
copied_dataset.to_csv("cleaned_data.csv")
```

```python
object=LabelEncoder()
df['location']=object.fit_transform(df['location'])
df.head()
```

```python
df.info()
```

```python
X=df.drop('price',axis=1)
y=df['price']
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=RandomForestRegressor(n_estimators=50,max_depth=5)
model.fit(Xtrain,ytrain)
```

```python
model.score(Xtest,ytest)
```

```python
model.score(Xtrain,ytrain)
```

```python
with open('RFmodel.pkl','wb') as f:
    pickle.dump(model,f)
```
